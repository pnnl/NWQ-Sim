```python
import stim
import time
import random

def benchmark_stim(n_qubits):
    for each in n_qubits:
        circuit = stim.Circuit()
        n_repeats = each

        for _ in range(n_repeats):
            target = random.randint(0, each - 1)
            circuit.append("H", [target])

            num = random.randint(1, each - 2)
            pos = random.choice([-1, 1])
            control = num
            target = num + pos
            circuit.append("CX", [control, target])

            circuit.append("S", [random.randint(0, each - 1)])

        simulator = stim.TableauSimulator()

        start = time.perf_counter()
        simulator.do_circuit(circuit)
        end = time.perf_counter()

        print("Time", end - start)

        filename = f"/Users/garn195/Project Repositories/NWQ-Sim/stabilizer/sim_bench/stim_{each}.txt"
        with open(filename, "w") as file:
            file.write("stim\n")
            file.write(f"{end - start}\n")
            file.write(f"{each}\n")

qubit_test = []
i = 2
while i < 16:
    qubit_test.append(i)
    i *= 2

benchmark_stim(qubit_test)
```

```python
import stim
import time
import random
n_qubits = []
n_repeats = 1
CXcount = 0
Scount = 0

for i in range(1, 101, 1):
    n_qubits.append(i*16)


for each in n_qubits:
    # Create the circuit once
    circuit = stim.Circuit()

    # Prepare the qubit range upfront
    qubits = range(each)

    # Append the bulk of operations in one go for all repeats

    circuit.append_operation('H', qubits[:-1])
    circuit.append_operation('S', qubits[:-1])


    circuit = circuit * n_repeats
    Hcount = each * n_repeats

    print(len(circuit))
    simulator = stim.TableauSimulator()

    start = time.perf_counter()
    simulator.do(circuit)
    end = time.perf_counter()

    sampler = circuit.compile_sampler()
    measurements = sampler.sample(shots=10)

    print("Time", end - start)
    print(measurements)

    filename = "/Users/garn195/Project Repositories/NWQ-Sim/stabilizer/sim_bench/stim2D_"+ (str)(each) + ".txt"
    with open(filename, "w") as file:
        file.write("stim2D\n")
        file.write((str)(end-start)+"\n")
        file.write((str)(each)+"\n")
        file.write((str)(Scount)+"\n")
        file.write((str)(Hcount)+"\n")
        file.write((str)(CXcount)+"\n")
```
