# Understand Match Graph

```python
import planar_z as pz # only X error can happen
import stim
import pymatching
```

Only insert X error.

```python
d = 3
rounds = 1
noise = 0.001
```

```python
noise_profile = [noise, 0, 0, 0]
sf_code = pz.PlanarSurfaceCode(d)
c_no_data = sf_code.build_full_surface_code_circuit(rounds, noise_profile, "Z", data_error=False)
c_no_data.diagram("matchgraph-3d")
```

1 round, only 1Q error. It happens (1) on the X syndrome qubit, after the first H or the second H.  
(1) The X error after the first H won't flip the final detectors because they propagate into an X stabilizer. The X error after the second H will flip the X syndrome qubit, which is not shown in the match graph because the first round detector only involves Z stabilizers.

```python
c = sf_code.build_full_surface_code_circuit(rounds, noise_profile, "Z") # with data qubit error before the round
c.diagram("matchgraph-3d")
```

Adding the 1Q X error happening before the first round will flip the measurement results in both rounds, so there is no vertical edges. However, these X errors remain on the data qubits and will flip the final round Z detectors. The 13 edges in the last layer correspond to 13 data qubits.

```python
noise_profile = [0, 0, noise, 0]
sf_code = pz.PlanarSurfaceCode(d)
c = sf_code.build_full_surface_code_circuit(rounds, noise_profile, "Z")
c.diagram("matchgraph-3d")
```

1 round, only measurement error. This error happens on (1) in the first round on the syndrome qubits, will flip the detector both in this round and the final round (vertical lines), and (2) in the final round on the data qubits, will flip one (on the boundary) or two (in the bulk) final round detectors.

```python
noise_profile = [0, 0, 0, noise]
sf_code = pz.PlanarSurfaceCode(d)
c = sf_code.build_full_surface_code_circuit(rounds, noise_profile, "Z")
c.diagram("matchgraph-3d")
```

1 round, only reset errors. (0) The reset error on the initialization of the data qubits behaves like the data qubit error before the round. As mentioned above, they will not give vertical edges but flip the final round detectors. (1) The reset error on the Z syndrome qubits will flip the Z detector in the first round, acting like a measurement error, which gives the vertical edges. These X errors will not propagate to the data qubits because they act on target qubits of CNOT.  
(2) The reset error on the X syndrome will become Z errors after H gate and will not propagate to the data qubits, and it becomes X error again before the measurement, flipping the X syndrome, but it's not shown in the graph since there's no X stabilizers included. (3) The reset error after syndrome qubit measurement will flip the both rounds of detectors and will not give vertical edges.

```python
noise_profile = [0, noise, 0, 0]
sf_code = pz.PlanarSurfaceCode(d)
c = sf_code.build_full_surface_code_circuit(rounds, noise_profile, "Z")
c.diagram("matchgraph-3d")
```

1 round, only X errors after CNOT. In the previous three cases, we can check that each edge has the same weight (each error mechanism happens with the same probability). This case is more complicated and we have edges with different weight. When inserting X error after CNOT, X error happens with equal probability = noise = 0.001 on each qubit.  

(1) The X error happening on the Z syndrome qubit behaves like a measurement error and will give vertical edges. The X error happening on the X syndrome qubit will propagate to data qubits, and if the X error weight is 1 or 3, the final round detector will be flipped. This gives the edges matched to the boundary.  
(2) The X error happening on the data qubits (during the syndrome extraction) will remain in the system and flip the final round detectors (1 or 2 depending on the error location).  
(3) The diagonal edges appears because the X errors happen both on the syndrome and data qubit after a CNOT. The X error on the syndrome flips the Z detector in the first round, and the X error on the data qubit is later checked by the adjacent Z stabilizer in the final measurement. That's why we see the diagonal edges. The reason that the diagonal edges only appear between some nodes is due to the gate scheduling. The X error on some data qubit after gate g can only flip the Z stabilizers that act on it later than g. (e.g. (1,4) to (3,4))
